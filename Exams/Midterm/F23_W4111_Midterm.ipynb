## W8

Briefly define and explain:

- Natural join
- Equi-join
- Theta join
- Self-join

<u>Answer</u>
Natural join: a type of join that creates a new resultant table by combining rows from two tables based on common values in their shared columns. The system automatically matches rows based on column names and data types that are same in both tables, and returns a combined table without repeating matching columns.

Equi-join: combines rows from two tables based on the equality condition specified in the join condition. It uses the equal sign as the comparison operator for matching rows based on values of specified columns.

Theta join: A type of join uses a comparison operator than the equal sign in its join condition. It relates rows within a single table based on a specified criterion, often using comparison operators such as < > <= >=. e.g. join two tables on a column value being greater in one than the other.

Self-join: A table is joined with itself. It is used to compare rows within same table based on specified relationship or condition between rows. It allows for comparison between rows of same dataset.

## Best Baseball Players

- This question uses ```lahmansbaseballdb.batting```, ```lahmansbaseballdb.pitching``` and ```lahmansbaseballdb.people```.<br><br>

- These query computes performance metrics:<br><br>
    - Batting:
        - On-base percentage: OBP is ```(sum(h) + sum(BB))/(sum(ab) + sum(BB))```
        - Slugging percentage: SLG is<br><br>

\dfrac{(sum(h) - sum(`2b`) - sum(`3b`) - sum(hr)) + 2*sum(`2b`) + 3*sum(`3b`) + 4*sum(hr))}

    - On-base percentage plus slugging: OPS is is ```(obp + slg)```.<br><br>

    - ```great_because```

    - ```op_slg``` is on-base percentage plus slugging.
    - ```total_abs``` is total at bats.
    - ```total_w``` is the total wins.
    - ```total_l``` is the total loses.
    - ```total_d``` is some of ```total_w``` and ```total_l```.
    - ```win_percentage``` is the winning percentage.
    
    
- __Note:__ Since I saved the query to Pandas and then read it back it,
    - All number become floating point. Your answer uses SQL and the values should be integers except for ```win_percentage.```
    - Pandas converts ```NULL``` to ```NaN```. In your SQL result, you will have ```None``` instead of ```NaN.```

````python
bb_greats = pd.read_csv('./baseball_greats.csv')
# bb_greats
````

<u>Answer</u>


````python
%%sql

WITH battingmetrics AS (
    SELECT
        playerid,
        (SUM(h) + SUM(bb))/(SUM(ab) + SUM(bb)) AS obp,
        ((SUM(h) - SUM(`2b`) - SUM(`3b`) - SUM(hr)) + 
        2*SUM(`2b`) + 3*SUM(`3b`) + 4*SUM(hr))/SUM(ab) AS slg,
        SUM(ab) AS total_abs
    FROM lahmansbaseballdb.batting
    GROUP BY playerid
    HAVING total_abs >= 3000 AND (obp + slg) >= 1.000
),
pitchingmetrics AS (
    SELECT
        playerid,
        SUM(w) AS total_w,
        SUM(l) AS total_l,
        SUM(w) * 1.0 / (SUM(w) + SUM(l)) AS win_percentage
    FROM lahmansbaseballdb.pitching
    GROUP BY playerid
    HAVING (total_w + total_l) >= 200 AND (win_percentage >= 0.70 OR total_w >= 300)
)
SELECT
    pe.playerid,
    pe.namelast AS namelast,
    pe.namefirst AS namefirst,
    CASE
        WHEN b.playerid IS NOT NULL THEN 'batter'
        WHEN pi.playerid IS NOT NULL THEN 'pitcher'
    END AS great_because,
    b.slg,
    b.obp,
    b.obp + b.slg AS op_slg,
    b.total_abs,
    pi.total_w,
    pi.total_l,
    pi.total_w + pi.total_l AS total_d,
    pi.win_percentage
FROM lahmansbaseballdb.people pe
LEFT JOIN battingmetrics b ON pe.playerid = b.playerid
LEFT JOIN pitchingmetrics pi ON pe.playerid = pi.playerid
WHERE b.playerid IS NOT NULL OR pi.playerid IS NOT NULL;
````

````python
%%sql
DROP TABLE  IF EXISTS countries;
DROP TABLE  IF EXISTS cities;

CREATE TABLE countries (
    iso3 CHAR(3) PRIMARY KEY,
    country VARCHAR(255) NOT NULL,
    iso2 CHAR(2) UNIQUE NOT NULL
);

CREATE TABLE cities (
    id SERIAL PRIMARY KEY,
    city VARCHAR(255) NOT NULL,
    city_ascii VARCHAR(255),
    lat DECIMAL(9,6) NOT NULL,
    lng DECIMAL(9,6) NOT NULL,
    iso3_fk CHAR(3) REFERENCES countries(iso3)
);

INSERT INTO countries (iso3, country, iso2)
SELECT DISTINCT iso3, country, iso2 
FROM F23W4111Midterm.worldcities;

INSERT INTO cities (id, city, city_ascii, lat, lng, iso3_fk)
SELECT id, city, city_ascii, lat, lng, iso3
FROM F23W4111Midterm.worldcities;
````
